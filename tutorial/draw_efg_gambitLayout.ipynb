```python
g = gbt.Game.new_tree(players=["Buyer", "Seller"], title="One-shot trust game, after Kreps (1990)")
g.append_move(g.root, player="Buyer", actions=["Trust", "Not trust"])
g.append_move(g.root.children["Trust"], player="Seller", actions=["Honor", "Abuse"])
g.set_outcome(g.root.children["Trust"].children["Honor"], outcome=g.add_outcome(payoffs=[1, 1], label="Trustworthy"))
g.set_outcome(g.root.children["Trust"].children["Abuse"], outcome=g.add_outcome(payoffs=[-1, 2], label="Untrustworthy"))
g.set_outcome(g.root.children["Not trust"], g.add_outcome(payoffs=[0, 0], label="Opt-out"))
draw_tree(g)
```
